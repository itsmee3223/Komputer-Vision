```python
# BOLA NOMOR 9
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from matplotlib.pyplot import figure

figure(figsize=(50, 50))
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Komputer Vision/gambar_uts/soal1/challenge2-detailed.jpg")
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
lower = np.array([166,109,86])
upper = np.array([173,199,195])
mask = cv2.inRange(hsv, lower, upper)
result = cv2.bitwise_and(image, image, mask=mask)

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```
